Deep Learning Model

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import accuracy_score
from numpy import mean
from numpy import std
```

```python
# x_train.shape
# y_train.shape
```

```python
def get_model(ip,op):
  model=Sequential()
  model.add(Dense(32,activation='relu',input_dim=ip,kernel_initializer='he_uniform'))
  model.add(Dense(op,activation='sigmoid'))
  model.compile(loss='binary_crossentropy',optimizer='adam')
  return model
```

```python



print(y.shape)
print(x.shape)
```

```python
from enum import auto
def eval_model(x,y):
  res=list()
  

  cv=RepeatedKFold(n_splits=10,n_repeats=10,random_state=442)
  
  for train,test in cv.split(x):
    x_train,x_test=x[train],x[test]
    y_train,y_test=y[train],y[test]
    
    model=get_model(19,1)
    model.fit(x_train,y_train,verbose=0,epochs=3)
    pred= model.predict(x_test)
    print(y_test)
    print(pred)
    pred=pred.round()
    print(pred)
    acc=accuracy_score(y_test,pred)

    print('>%.3f'%acc)
    res.append(acc)
  return res

res=eval_model(x,y)

print('Accuracy: %.3f (%.3f)' % (mean(res), std(res)))
```
